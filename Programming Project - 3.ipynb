# Project 3 - Rock, Paper, Scissors Game

```python
import random

comp_wins = 0
player_wins = 0

def Player_Option():
    user_choice = input("Choose Rock, Paper or Scissors = ").lower()
    if user_choice in ['rock', 'r']:
        user_choice = "r"
    elif user_choice in ['paper', 'p']:
        user_choice = "p"
    elif user_choice in ['scissors', 's']:
        user_choice = "s"
    else:
        print("Invalid. Please try again")
    return user_choice

def Computer_Option():
    comp_choice = random.randint(1,3)
    if comp_choice == 1:
        comp_choice = "r"
    elif comp_choice == 2:
        comp_choice = "p"
    else:
        comp_choice = "s"
    return comp_choice

while True:
    user_choice = Player_Option()
    comp_choice = Computer_Option()
    if user_choice == "r":
        if comp_choice == "r":
            print("Computer chose rock. You tied.")
        elif comp_choice == "p":
            print("Computer chose paper. You lose.")
            comp_wins += 1
        elif comp_choice == "s":
            print("Computer chose scissors. Congrats, you win.")
            player_wins += 1
            
    elif user_choice == "p":
        if comp_choice == "p":
            print("Computer chose paper. You tied.")
        elif comp_choice == "r":
            print("Computer chose rock. Congrats, You win.")
            player_wins += 1
        elif comp_choice == "s":
            print("Computer chose scissors. You lose.")
            comp_wins += 1
            
    elif user_choice == "s":
        if comp_choice == "s":
            print("Computer chose scissors. You tied.")
        elif comp_choice == "p":
            print("Computer chose paper. Congrats, You win.")
            player_wins += 1
        elif comp_choice == "r":
            print("Computer chose rock. You lose.")
            comp_wins += 1
    print()
    print("Player wins = ",player_wins)
    print("Computer wins = ",comp_wins)
    print()
    
    user_choice = input("Do you want to continue? ").lower()
    if user_choice in ['yes', 'y']:
        pass
    elif user_choice in ['no', 'n']:
        print("Game ended. See you again")
        break
    else:
        break
```
